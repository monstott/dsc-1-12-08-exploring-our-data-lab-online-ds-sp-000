### Build normalized histograms with kde plots to explore the distributions further. 
### Use only the continuous variables in the dataset to plot these visualizations. 

```python
#Â Your code here 
for column in ['Size', 'Temperature', 'Fuel_Price', 'CPI', 'Unemployment', 'Weekly_Sales']:
    walmart[column].plot.hist(normed=True)
    walmart[column].plot.kde(label=column+" kde")
    plt.legend()
    plt.show()
```

```python
# State your observations here 
#A few of these distributions contain multiple peaks. The multimodal columns are CPI and Size.
#Adjustment to binning are required to a few columns is recommended, since little information is provided by the above
#distributions. This is especially true for Weekly_Sales, and to a lesser extent for CPI and Size.
```

### Build joint plots to check for the linearity assumption between predictors and target variable

Let's use a slightly more advanced plotting technique in seaborn that uses scatter plots, distributions, kde and simple regression line - all in a single go. Its called a `jointplot`. [Here is the official doc. for this method](https://seaborn.pydata.org/generated/seaborn.jointplot.html). 

Here is how you would use it:

> **`sns.jointplot(x= <column>, y= <column>, data=<dataset>, kind='reg')`**

A joint plot will allow us to visually inspect linearity as well as normality assumptions as a single step. 

```python
#Â Your code here 
for column in ['Store', 'Dept', 'Size', 'Temperature', 'Fuel_Price', 'CPI', 'Unemployment']:
    sns.jointplot(x=column,y="Weekly_Sales",data=walmart, kind='reg',label=column,joint_kws={'line_kws':{'color':'green'}})
    plt.legend()
    plt.show()
```

```python
#Â Provide your observations here 
#A few of the distributions above follow a uniform distribution. This includes Fuel_Price and Temperature. 
#Other columns have discrete column values available. This includes CPI, Size, and Store.
#The remaining columns reside inbetween. This includes Unemployment and Dept.
#The normality assumption does not hold for the predictors of Weekly_Sales.
#From the simple linear regression lines, the variables Store, Dept, Fuel_Price, Temperature, and Unemployment appear
#flat with changes in Weekly_Sales. These variables lack a relationship with Weekly_Sales.
#The slope of the CPI simple regression line is slightly negative, indicating a greater relationship with Weekly_Sales than
#the columns listed as having flat slopes.
#The slope of the Size simple regression line is slightly positive, indicating a greater relationship with Weekly_Sales than
#the columns listed as having flat slopes.
# Size of the store has a weak positive realtionship with sales
```
